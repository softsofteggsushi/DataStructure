```python
import sys

def prim(adj_list, s):
    N = len(adj_list)
    visited = [False] * N
    D = [sys.maxsize] * N
    D[s] = 0
    previous = [None] * N
    previous[s] = s

    for _ in range(N):
        m = -1
        min_value = sys.maxsize
        for j in range(N):
            if not visited[j] and D[j] < min_value:
                min_value = D[j]
                m = j
        visited[m] = True

        for w, wt in adj_list[m]:
            if not visited[w]:
                if wt < D[w]:
                    D[w] = wt
                    previous[w] = m

    return previous, D
```

```python

# 가중치 그래프의 연결 리스트
adj_list = {
    0: [(1, 2), (2, 1)],
    1: [(0, 2), (3, 3), (4, 4)],
    2: [(0, 1), (5, 5)],
    3: [(1, 3), (6, 6)],
    4: [(1, 4)],
    5: [(2, 5)],
    6: [(3, 6), (7, 7)],
    7: [(6, 7)]
}

start_vertex = 0

previous, D = prim(adj_list, start_vertex)

# 결과 출력
print("Previous:", previous)
print("D:", D)
```

```python
def find(p, u):
    if u != p[u]:
        p[u] = find(p, p[u])  # 경로 압축
    return p[u]

def union(p, u, v):
    root1 = find(p, u)
    root2 = find(p, v)
    p[root2] = root1

def kruskal(adj_list, weights):
    N = len(adj_list)
    p = [i for i in range(N)]
    mst = []
    tree_edges = 0
    mst_cost = 0

    while True:
        if tree_edges == N - 1:
            break

        u, v, wt = weights.pop(0)  # 다음 최소 가중치 간선 가져오기

        if find(p, u) != find(p, v):
            union(p, u, v)
            mst.append((u, v))  # 트리에 (u, v) 추가
            mst_cost += wt
            tree_edges += 1

    return mst, mst_cost

```

```python
# 가중치 그래프의 연결 리스트
adj_list = {
    0: [(1, 2), (2, 1)],
    1: [(0, 2), (3, 3), (4, 4)],
    2: [(0, 1), (5, 5)],
    3: [(1, 3), (6, 6)],
    4: [(1, 4)],
    5: [(2, 5)],
    6: [(3, 6), (7, 7)],
    7: [(6, 7)]
}

# 가중치 정보
weights = []
for u in adj_list:
    for v, wt in adj_list[u]:
        if u < v:  # 중복된 간선은 제외하여 weights에 추가
            weights.append((u, v, wt))
weights.sort(key=lambda x: x[2])  # 가중치 기준으로 정렬

mst, mst_cost = kruskal(adj_list, weights)

# 결과 출력
print("Minimum Spanning Tree:")
for u, v in mst:
    print(u, "-", v)
print("Total Cost:", mst_cost)
```
