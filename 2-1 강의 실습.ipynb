#**제어문:조건문**

**if**

```python
#True
number = 10

if number > 0:
    print('양수입니다.')

print('조건문 밖입니다')
```

```python
#False
number = -5

if number > 0:
    print('양수입니다.')

print('조건문 밖입니다')
```

**else if**

```python
number = 10

if number > 0:
    print('양수입니다')

else:
    print('음수입니다')

print('조건문 밖입니다')
```

```python
number = 0

if number > 0:
    print("양수입니다")

elif number == 0:
    print('영입니다')
else:
    print('음수입니다')

print('조건문 밖입니다')
```

# **제어문:반복문**

```python
#for
for i in range(4):
    print(i)
```

```python
#while
i = 0
n = 4

while i < n:
    print(i)
    i = i + 1
```

반복문:continue & break

```python
#break
i = 1
while (i<=10):
    print('6 * ',(i), '=',6 * i)
    if i >= 5:
        break
    i = i + 1
```

```python
#continue
#(셀의 일시정지 버튼으로 강제종료)
i = 1
while (i<=10):
    print('6 * ',(i), '=',6 * i)
    if i >= 5:
        continue
    i = i + 1

```
