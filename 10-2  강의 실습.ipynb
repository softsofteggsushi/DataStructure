```python
class Node:
    def __init__(self, key, value, left=None, right=None):
      self.key = key
      self.value = value
      self.left = left
      self.right = right
      
class BinaryTree:
    def __init__(self):                                   # tree constructor
      self._root = None
    def search(self,t, k):
      if t == None:
        return None                                       # key를 발견 못함
      if t.key > k:                                       # 왼쪽 서브 트리 탐색
       return self.search(t.left, k)
      elif t.key < k:                                     # 오른쪽 서브 트리 탐색
        return self.search(t.right, k)
      else:
        return t.value                                     # key를 가진 노드 발견
    def insert(self, key, value):                          # 삽입 연산
      self.root = self.insert_item(self.root, key, value)
    def insert_item(self, t, key, value):
      if t == None:
       return Node(key, value)
      if t.key > key:                                      # 왼쪽 서브 트리에 삽입
       t.left = self.insert_item(t.left, key, value)
      elif t.key < key:                                    # 오른쪽 서브 트리에 삽입
        t.right = self.insert_item(t.right, key, value)
      else:                                                # 노드 n의 value 갱신
        t.vlaue = value
      return t
    def levelorder(self): # 레벨 순회
      q = []
      n = self.root
      q.append(n)
      while len(q) != 0:
        t = q.pop(0)
        print(str(t.key), ' ', end='')
        if t.left != None:
          q.append(t.left)
        if t.right != None:
          q.append(t.right)   
    def min(self):                                         # 최솟값 가진 노드 찾기
     if self.root == None:
      return None
     return self.minimum(self.root) 
    def minimum(self, t):
     if t.left == None:
      return t
     return self.minimum(t.left)
    def delete_min(self):                                 # 최소값 삭제
      if self.root == None:
        print('Tree is empty')
      self.root = self.del_min(self.root)
    def del_min(self, t):
      if t.left == None:
        return t.right                                    # t의 오른쪽 자식 반환
      t.left = self.del_min(t.left)                       # t의 왼쪽 자식으로 순환 호출
      return t
    def del_node(self, t, k):
      if t == None:
        return None
      if t.key > k: # 왼쪽으로 이동
        t.left = self.del_node(t.left, k)
      elif t.key < k: # 오른쪽으로 이동
        t.right = self.del_node(t.right, k)
      else: # 삭제할 노드 발견
        if t.right == None: # 자식이 0개 or 1개
          return t.left
        if t.left == None: # 자식이 0개 or 1개
          return t.right
        target = t # 자식이 2개
        n = self.minimum(target.right)
        t.right = self.del_min(target.right)
        t.left = target.left
      return t         
         
```
