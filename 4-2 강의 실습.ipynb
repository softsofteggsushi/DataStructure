```python
#init list node
class ListNode:
  def __init__(self):
    self.data = None
    self.link = None

memory = []
head,current,prev = None,None,None

```

```python
#print Circlar Linked list
def print_Clist(node):
  head = node
  current = node
  while(current.link != None):
    print(current.data,end = "->")
    current = current.link
    if(current == head):
      break;
  print(current.data,"...(iteration)")
```

```python
node0 = ListNode()
node0.data = 0

head = node0
node0.link = head
memory.append(node0)
print(memory[0].data)
```

```python
prev = node0
node1 = ListNode()
node1.data = 10

prev.link = node1
node1.link = head

memory.append(node1)
print(memory[1].data)
print_Clist(head)
```

```python
#insert in front
node = ListNode()
node.data = 50
node.link = head
last = head


while (last.link != head):
  last = last.link

last.link = node
head = node

print_Clist(head)

```

```python
#insert 70 before 10 

node = ListNode()
node.data = 70
before = 10
current = head

while current.link != head:
  prev = current
  current = current.link
  if current.data == before:
      node.link = current
      prev.link = node

print_Clist(head)
```

```python
#insert at endnode

node4 = ListNode()
node4.data = 30
node4.link = head
last = head

while last.link != head:
  last = last.link

last.link = node4

print_Clist(head)
```

```python
#delete first


current = head
head = head.link
last = head

while last.link != current:
  last = last.link
last.link = head
del(current)

print_Clist(head)
```
