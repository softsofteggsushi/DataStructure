```python
class HashTable:
    def __init__(self, size):
        self.M = size               #hash table의 크기
        self.a = [None] * size      #key를 저장하는 배열
        self.d = [None] * size      #data를 저장하는 배열

    def hash(self, key):      #hash function
        return key % self.M   #key 값을이용하여 테이블 내의 위치 매핑

    def put(self, key, data): # 삽입 연산
        initial_position = self.hash(key) # 초기 위치
        i = initial_position
        j = 0
        while True:
            if self.a[i] is None or self.a[i] == '$': # 삽입 위치 발견
                self.a[i] = key # key를 해시 테이블에 저장
                self.d[i] = data # key관련 데이터 저장
                return
            if self.a[i] == key: # 이미 key 존재하면
                self.d[i] = data # 데이터만 갱신
                return
            j += 1
            i = (initial_position + j) % self.M # i의 다음 위치
            if i == initial_position: # i가 초기 위치와 같으면 루프 종료
                break

    def delete(self, key): # 삭제 연산
        initial_position = self.hash(key)
        i = initial_position
        j = 1
        while self.a[i] is not None or self.a[i] == '$': # a[i]가 empty가 아니면
            if self.a[i] == key:
                self.a[i] = '$' # '$'로 삭제된 곳 표시
                return
            j += 1
            i = (initial_position + j) % self.M # i의 다음 위치
        print("삭제하려는", key, "가(이) 테이블에 미존재") # 삭제 실패

    def get(self, key): # 탐색 연산
        initial_position = self.hash(key)
        i = initial_position
        j = 1
        while self.a[i] is not None: # a[i]가 empty가 아니면
            if self.a[i] == key:
                return self.d[i] # 탐색 성공
            i = (initial_position + j) % self.M # i의 다음 위치
            j += 1
            if i == initial_position: # i가 초기 위치와 같으면 루프 종료
                return None # 탐색 실패
        return None # 탐색 실패
```

```python
hash_table = HashTable(10)
hash_table.put(5, "Data 1")
hash_table.put(15, "Data 2")
hash_table.put(25, "Data 3")

print(hash_table.get(5))  # Output: Data 1
print(hash_table.get(15))  # Output: Data 2
print(hash_table.get(25))  # Output: Data 3

hash_table.delete(15)
print(hash_table.get(15))  # Output: None (삭제된 데이터)
```
