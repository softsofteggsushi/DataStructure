```python
class Node:
  def __init__(self, key : int ,left=None, right=None): # Node constructor
    self.key = key
    self.left = left
    self.right = right
class BinaryTree:
  def __init__(self): # tree constructor
    self._root = None
  def preorder(self, n): # 전위 순회
    if n != None:
      print(str(n.key),' ', end='')
      if n.left:
        self.preorder(n.left)
      if n.right:
        self.preorder(n.right)
  def inorder(self, n): # 중위 순회
    if n != None:
      if n.left:
        self.inorder(n.left)
      print(str(n.key),' ', end='')
      if n.right:
        self.inorder(n.right)
  def postorder(self, n): # 후위 순회
    if n != None:
      if n.left:
        self.postorder(n.left)
    if n.right:
      self.postorder(n.right)
    print(str(n.key),' ', end='')
  def levelorder(self, n): # 레벨 순회
    q = []
    q.append(n)
    while len(q) != 0:
      t = q.pop(0)
      print(str(t.key), ' ', end='')
      if t.left != None:
        q.append(t.left)
      if t.right != None:
        q.append(t.right)
  def size(self, root): #트리의 노드 수 계산
    if root is None:
      return 0
    else:
      return 1 + self.size(root.left) + self.size(root.right)
  def height(self, root): # 트리의 높이 계산
    if root == None:
      return 0
    return max(self.height(root.left),self.height(root.right)) + 1
```

```python
t = BinaryTree() # 이진 트리 객체 t 생성
n1 = Node(5)     
n2 = Node(3)
n3 = Node(9)
n4 = Node(1)
n5 = Node(4)
n6 = Node(6)

n1.left = n2  # n1의 왼쪽 자식-> n2
n1.right = n3 # n1의 오른쪽 자식-> n3
n2.left = n4  # n2의 왼쪽 자식-> n4
n2.right = n5 # n2의 오른쪽 자식-> n5
n3.left = n6  # n3의 왼쪽 자식-> n6
 
t.root = n1   # t의 루트를 n1으로
```

```python
print("%5s"%t.root.key)
print("%4s"%t.root.left.key,t.root.right.key)
print("%2s"%t.root.left.left.key,t.root.left.right.key,t.root.right.left.key)
```

#Thread Binary Tree

```python
class Node:
  def __init__(self, key , left=None, right=None, rightThread=False):
  # Node constructor
    self.key = key
    self.left = left
    self.right = right
    self.rightThread = rightThread

class ThreadBinaryTree:
  def __init__(self): # tree constructor
    self._root = None
  def leftMost(self, n):
    if (n == None):
      return None
    while (n.left != None):
      n = n.left
    return n
  def inorder(self, root):
    cur = self.leftMost(root);
    while (cur != None):
      print(cur.key, " ", end='')
      if (cur.rightThread):
        cur = cur.rightThread
      else: # Else go to the leftmost child in right
      # subtree
        cur = self.leftMost(cur.right)
```

```python
T = ThreadBinaryTree()
n1 = Node("A")     
n2 = Node("B")
n3 = Node("E")
n4 = Node("C")
n5 = Node("D")
n6 = Node("F")
n7 = Node("G")

n1.left = n2  # n1의 왼쪽 자식-> n2
n1.right = n3 # n1의 오른쪽 자식-> n3
n2.left = n4  # n2의 왼쪽 자식-> n4
n2.right = n5 # n2의 오른쪽 자식-> n5
n3.left = n6  # n3의 왼쪽 자식-> n6
n3.right = n7
n4.rightThread = n2
n5.rightThread = n1
n6.rightThread = n3
 
T.root = n1   # t의 루트를 n1으로
```

```python
print("%5s"%T.root.key)
print("%4s"%T.root.left.key,T.root.right.key)
print("%2s"%T.root.left.left.key,T.root.left.right.key,T.root.right.left.key,T.root.right.right.key)
```
