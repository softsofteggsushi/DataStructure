```python
import math
import time
```

```python
#삽입 정렬
def insertionSort(A):
  for i in range(1, len(A)):
    loc = i-1
    newItem = A[i]
    while loc >= 0 and newItem < A[loc]:
      A[loc+1] = A[loc]
      loc -= 1
    A[loc+1] = newItem
```

```python
#병합 정렬
def mergeSort(A, p:int, r:int):             
  if p < r:
    q = (p+r) // 2
    mergeSort(A, p, q)
    mergeSort(A, q+1, r)
    merge(A, p, q, r)
def merge(A, p:int, q:int, r:int):
  i = p; j = q+1; t = 0
  tmp = [0 for i in range(len(A))]
  while i <= q and j <= r:
    if A[i] <= A[j]:
      tmp[t] = A[i]; t += 1; i += 1
    else:
     tmp[t] = A[j]; t += 1; j += 1
  while i <= q:
   tmp[t] = A[i]; t += 1; i += 1
  while j <= r:
   tmp[t] = A[j]; t += 1; j += 1
  i = p; t = 0
  while i <= r:
    A[i] = tmp[t]; t += 1; i += 1
```

```python
#버블 정렬
def bubbleSort(A):
  for numElements in range(len(A), 0, -1):
    for i in range(numElements-1):
      if A[i] > A[i+1]:
        A[i], A[i+1] = A[i+1], A[i]  
```

```python
#선택 정렬
def selectionSort(A):
  for last in range(len(A)-1, 0, -1):
    k = theLargest(A, last)
    # A[0...last] 중 가장 큰 수 A[k] 찾기
    A[k], A[last] = A[last], A[k]

def theLargest(A, last:int) -> int:
  # A[0...last]에서 가장 큰 수의 인덱스를 리턴한다
  largest = 0
  for i in range(last+1):
    if A[i] > A[largest]:
      largest = i
  return largest
```

```python
#삽입 정렬
arr = [3, 31, 48, 73, 8,33, 11, 15, 20, 29, 28, 65, 25, 4,12]
start = time.time()
insertionSort(arr)
end = time.time()

print(arr)
print("걸린시간 :"+ f"{end - start:.5f} sec")
```

```python
#병합 정렬
arr = [3, 31, 48, 73, 8,33, 11, 15, 20, 29, 28, 65, 25, 4,12]
start = time.time()
mergeSort(arr,0,len(arr)-1)
end = time.time()

print(arr)
print("걸린시간 :"+ f"{end - start:.5f} sec")
```
