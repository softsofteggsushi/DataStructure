```python
#Local scope example
def local_scope():
  local_var = 1
  print(local_var)

#print(local_var) #local_var is not defined(out of local_scope)

local_scope()
```

```python
#Enclosed scope example
def enclosing_scope():
  enclosing_var = 2
  def inner_func():
    inner_var = 3
    print(enclosing_var * inner_var)
  inner_func()
  #print(inner_var) #inner_var is not defined (out of inner_func local scope)

enclosing_scope()
```

```python
#Module scope example
global_var = 10
def outer_func():
  outer_var = global_var + 3
  print(outer_var)
  def inner_func():
    inner_var = global_var * 10
    print(inner_var)
  inner_func()
outer_func()
print(global_var)
```
