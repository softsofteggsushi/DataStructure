```python
#init list node
class ListNode:
  def __init__(self):
    self.data = None
    self.link = None


node0 = ListNode()
node0.data = 0
head = node0
print(node0.data)
```

```python
#print list
def print_Clist(node):
  head = node
  current = node
  while(current.link != None):
    print(current.data,end = "->")
    current = current.link
    if(current == head):
      break;
  print(current.data)
```

```python
#link 10 node

node1 = ListNode() #init node1
node1.data =10     #store data

node0.link = node1 # 0 -> 10
node1.link = node0 # 10 -> 0


print(node0.data) 
print(node0.link.data)
print(node0.link.link.data)
```

```python
node2 = ListNode()
node2.data = 20
node1.link = node2  #0 -> 10 -> 20
node2.link = node0  #0 -> 10 -> 20 -> 0(head)

node3 = ListNode()
node3.data = 30
node2.link = node3
node3.link = node0

node4 = ListNode()
node4.data = 40
node3.link = node4
node4.link = node0


print(node0.data)
print(node0.link.data)
print(node0.link.link.data)
print(node0.link.link.link.data)
print(node0.link.link.link.link.data)
print(node0.link.link.link.link.link.data)
print_Clist(node0)
```

```python
#insert 50 between 20 and 30
node5 = ListNode()
node5.data = 50
node5.link = node2.link
node2.link = node5

print_Clist(node0)
```

```python
#delete 20
node1.link = node2.link
del(node2)

print_Clist(node0)
```
