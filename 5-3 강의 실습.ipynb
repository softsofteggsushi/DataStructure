5-3 강의 실습

```python
class LListNode:
    def __init__(self, data):
        self.data = data
        self.link = None

class LListStack():
    def __init__(self):
        self.top = None
        
    def push(self, data):
        new_node = LListNode(data)
        new_node.link = self.top
        self.top = new_node
        
    def pop(self):
        pop_value = None
        if self.isStackEmpty():
            print("Stack is Empty")
        else:
            pop_data = self.top.data
            self.top = self.top.link
        return pop_data
    
    def peek(self):
        top_value = None
        if self.isStackEmpty():
            print("Stack is Empty")
        else:
            top_value = self.top.data            
        return top_value
    
    def isStackEmpty(self):
        if self.top is None:
            return True
        else:
          return False
    def printStack(self):
        print("---------stack status---------")
        head = self.top
        current = self.top
        while(current.link != None):
          print(current.data,end = "\n")
          current = current.link
          if(current == head):
            break;
        print(current.data)
```

```python
#Push

University = LListStack()

University.push("ChungAng")
University.push("Hanyang")
University.push("Sungkyunkwan")
University.push("Sogang")
University.push("Yonsei")
University.push("Seoul")

University.printStack()
```

```python
#Pop

Univ= University.pop()
print(Univ)
Univ = University.pop()
print(Univ)

University.printStack()
```

```python
#Peek

showUniv = University.peek()
print(showUniv)
showUniv = University.peek()
print(showUniv)

University.printStack()
```

#reverse string 

```python
def reverse(str):
  charStack = LListStack()
  for i in range(len(str)):
    charStack.push(str[i])
  out = ""
  while (charStack.isStackEmpty() == False):
    out += charStack.pop()
  return out

input = "Yohan is a teacher!"
answer = reverse(input)
print("Input string: ", input)
print("Reversed string: ", answer)
```
