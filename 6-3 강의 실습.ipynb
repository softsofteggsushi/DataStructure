```python
class CircleQueue():
    def __init__(self,SIZE):
        self.SIZE = SIZE
        self.queue = [ None for i in range(self.SIZE)]
        self.front = self.rear = 0

    def enQueue(self,data) :
       if (self.isQueueFull()) :
          print("Queue is full")
          return
       self.rear = (self.rear + 1) % self.SIZE
       self.queue[self.rear] = data

    def deQueue(self) :
       if (self.isQueueEmpty()) :
          print("Queue is empty")
          return None
       self.front = (self.front + 1) % self.SIZE
       data = self.queue[self.front]
       self.queue[self.front] = None
       return data
         
    def peek(self):
       if(self.isQueueEmpty()) :
          print("Queue is empty")
          return None
       return self.queue[self.front+1]

    def isQueueFull(self) :
        if ((self.rear + 1) % self.SIZE == self.front) :
          return True
        else: 
          return False
    def isQueueEmpty(self) :
        if (self.front == self.rear) :
          return True
        else :
          return False

    def printQueue(self):
      print("----- Queue status -----")
      for i in range(0, len(self.queue), 1) :
        print(self.queue[i], end=' ')
```

```python
Colors = CircleQueue(5)

#full queue
Colors.enQueue("Yellow")
Colors.enQueue("Green")
Colors.enQueue("Blue")
Colors.enQueue("Red")
Colors.enQueue("Pink")  #is full
Colors.printQueue()
print("Colors queue is Full? ->",Colors.isQueueFull())

print("\n")

#2 dequeue
Colors.deQueue();
Colors.deQueue();
Colors.printQueue()
print("Colors queue is Full? ->",Colors.isQueueFull()) #not Full but true

print("\n")

#insert
Colors.enQueue("Purple")
Colors.enQueue("Skyblue")
Colors.printQueue()

print("\n")
#delete
Colors.deQueue();
Colors.deQueue();
Colors.printQueue()
```
