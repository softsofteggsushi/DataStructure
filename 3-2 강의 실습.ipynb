3-2 강의 실습

```python
class ListNode:
  def __init__(self):
    self.data = None
    self.link = None

node0 = ListNode()
node0.data = 0
print(node0.data)
```

```python
node1 = ListNode()
node1.data = 10
node0.link = node1
print(node1.data)

# 0 -> 10
```

```python
node2 = ListNode()
node2.data = 20
node1.link = node2
print(node2.data)

node3 = ListNode()
node3.data = 30
node2.link = node3
print(node3.data)

node4 = ListNode()
node4.data = 40
node3.link = node4
print(node4.data)


#node0->node1->node2->node3->node4
#0->10->20->30->40

```

```python
current = node0
while(current.link != None):
  print(current.data)
  current = current.link

print("------------link print-----------")
print(node0.data)
print(node0.link.data)
print(node0.link.link.data)
print(node0.link.link.link.data)
print(node0.link.link.link.link.data)
```

```python
#insert
new_node = ListNode()           #new node
new_node.data = 50              #new node data
new_node.link = node3           #new node link 
node2.link = new_node           #insert


print(node0.data)
print(node0.link.data)
print(node0.link.link.data)
print(node0.link.link.link.data)
print(node0.link.link.link.link.data)
print(node0.link.link.link.link.link.data)

#0->10->20->50->30->40 insert between 20 and 30
```

```python
#delete
node1.link = node3            #node1 -> node 3
del(node2)                    #delete node2


print(node0.data)             #node0
print(node0.link.data)        #node1
print(node0.link.link.data)   #node3
```
