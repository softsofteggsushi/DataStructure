```python
import math
```

```python
def countingSort(A):
    k = max(A)
    C = [0 for _ in range(k + 1)]
    
    for j in range(len(A)):
        C[A[j]] += 1
    
    for i in range(1, k + 1):
        C[i] = C[i] + C[i - 1]
    
    B = [0 for _ in range(len(A))]
    for j in range(len(A) - 1, -1, -1):
        B[C[A[j]] - 1] = A[j]
        C[A[j]] -= 1
    
    return B


def radixSort(A):
    maxValue = max(A)
    numDigits = math.ceil(math.log10(maxValue))
    # 자릿수 계산
    bucket = [[] for _ in range(10)]
    # 0, 1, ..., 9에 대한 10개의 리스트 (10진수이므로)
    
    for i in range(numDigits):
        # 각 자리수에 대한 반복문
        for x in A:
            y = (x // 10**i) % 10
            # A 배열 숫자들의 각 자리수만 비교
            bucket[y].append(x)
        
        A.clear()
        for j in range(10):
            A.extend(bucket[j])
            bucket[j].clear()

    return A

def bucketSort(A):
    n = len(A)
    B = [[] for _ in range(n)]
    
    for i in range(n):
        B[math.floor(n * A[i])].append(A[i])
    
    A.clear()
    for i in range(n):
        insertionSort(B[i])
        A.extend(B[i])
    
    return A

def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

```

```python
# 계수 정렬 알고리즘 예제
A = [4, 2, 2, 8, 3, 3, 1]
result = countingSort(A)

# 결과 출력
print(result)
```

```python
#기수 정렬 알고리즘 예제
A = [170, 45, 75, 90, 802, 24, 2, 66]
result = radixSort(A)

# 결과 출력
print(result)
```

```python
# 버킷 정렬 알고리즘 예제
A = [0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.12, 0.23, 0.68]
result = bucketSort(A)

# 결과 출력
print(result)
```
