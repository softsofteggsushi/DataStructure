#Class Stack

```python
class Stack():
    def __init__(self,stackSize):
        self.SIZE = stackSize
        self.stack = [ None for i in range(self.SIZE)]
        self.top = -1
    def push(self, data):
      if(self.isStackFull()) :
        print("Stack is Full")
        return
      self.top += 1
      self.stack[self.top] = data
        
    def pop(self):
        pop_value = None
        if self.isStackEmpty():
            print("Stack is Empty")
            return None
        else:
            pop_value = self.stack[self.top]
            self.stack[self.top] = None
            self.top -= 1
        return pop_value
            
    def peek(self):
        peek_vaule = None
        if self.isStackEmpty():
            print("Stack is Empty")
        else:
            peek_value = self.stack[self.top]       
        return peek_value

    def isStackFull(self) :
      if (self.top >= self.SIZE - 1) :
          return True
      else :
          return False
         
    def isStackEmpty(self):      
        if self.top  == -1:
            return True
        else :
          return False
    def printStack(self):
      print("-------stack status-------")
      for i in range(len(self.stack)-1,-1,-1):
        print(self.stack[i])
```

```python
#Push
Champions = Stack(10)

Champions.push("Riven")
Champions.push("Aatrox")
Champions.push("Annie")
Champions.push("Aurelion sol")
Champions.push("Ashe")
Champions.push("Faker")
Champions.push("Keria")


Champions.printStack()
```
