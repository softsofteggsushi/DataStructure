```python
class heap:
  def __init__(self):
    self.__A = []
  def insert(self, x):
    self.__A.append(x)
    self.__percolateUp(len(self.__A)-1)

  def __percolateUp(self, i:int):                                   # Circular call percolateUp
    parent = (i - 1) // 2
    if i > 0 and self.__A[i] > self.__A[parent]:
      self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
      self.__percolateUp(parent)

  def deleteMax(self):                                                                  
    if (not self.isEmpty()):                                        # heap is in self.__A[0...len(self.__A)-1]
      max = self.__A[0]
      self.__A[0] = self.__A.pop()                                  # *.pop(): 리스트의 끝원소 삭제 후 원소 리턴
      self.__percolateDown(0)
      return max
    else:
      return None
  def __percolateDown(self, i:int):                                 # Circular call percolateDown
    child = 2 * i + 1                                               # left child
    right = 2 * i + 2                                               # right child
    if (child <= len(self.__A)-1):
      if (right <= len(self.__A)-1 and self.__A[child]< self.__A[right]):
        child = right                                               # index of larger child
      if self.__A[i] < self.__A[child]:
        self.__A[i], self.__A[child] = self.__A[child],self.__A[i]
        self.__percolateDown(child)                                                
  def isEmpty(self):
    if not self.__A:
      return True
    if self.__A:
      return False

  def buildHeap(self):
    for i in range((len(self.__A) - 2) // 2, -1, -1):
      self.__percolateDown(i)

  def printHeap(self):
    print(self.__A)
```

```python
BHeap = heap()

BHeap.insert(9)
BHeap.insert(6)
BHeap.insert(8)
BHeap.insert(4)
BHeap.insert(3)
BHeap.insert(3)
BHeap.insert(5)
BHeap.insert(2)
BHeap.insert(1)

BHeap.printHeap()
```
