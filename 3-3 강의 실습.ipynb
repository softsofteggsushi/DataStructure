```python
#init list node
class ListNode:
  def __init__(self):
    self.data = None
    self.link = None

memory = []
head, current, prev = None, None, None


node0 = ListNode()
node0.data = 0
head = node0
print(node0.data)
```

```python
#print list
def print_list(node):
  current = node
  while(current.link != None):
    print(current.data,end = "->")
    current = current.link
  print(current.data)
```

```python

node1 = ListNode()
node1.data = 10
node0.link = node1

node2 = ListNode()
node2.data = 20
node1.link = node2

node3 = ListNode()
node3.data = 30
node2.link = node3


node4 = ListNode()
node4.data = 40
node3.link = node4

head = node0

prev = head.link
current = prev.link


print("head node data : ", head.data)
print("prev node data : ", prev.data)
print("current node data : ", current.data)
print_list(head)

```

```python
#insert into first node
node = ListNode()
node.data = 50
node.link = head
head = node

print("head node data : ", head.data)
print_list(head)
```

```python
#insert 60 between 20 and 30
current = head
find_data = 30
insert_data = 60

while current.data != find_data :
  prev = current
  current = current.link

insert_node = ListNode()
insert_node.data = insert_data
insert_node.link = current
prev.link = insert_node

print_list(head)
```

```python
#delete first node
current = head
head = head.link
del(current)

print_list(head)
```
