# set

```python
s1 = set([1,2,3])
print(s1)
s2 = set("Hello")
print(s2) #중복 x
```

```python
#set 연산
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print (s1 & s2)               #교집합
print (s1.intersection(s2))   
print (s1 | s2)               #합집합
print (s1.union(s2))          
print(s1 - s2)                #차집합
print(s1.difference(s2))
print(s2 - s1)
print(s2.difference(s1))
```

# variable

```python
a = [1,2,3,]
b = a

print("-----------a[1] = 4 변경 전----------")
print(a)
print(b)

a[1] = 4

print("-----------a[1] = 4 변경 후----------")
print(a)
print(b)


print("---------------주소값----------------")
print(id(a))
print(id(b))
```

```python
a = [1,2,3,]
b = a [:]

print("-----------a[1] = 4 변경 전----------")
print(a)
print(b)

a[1] = 4

print("-----------a[1] = 4 변경 후----------")
print(a)
print(b)


print("---------------주소값----------------")
print(id(a))
print(id(b))
```
