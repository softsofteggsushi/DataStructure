```python
adj_list = [[1], [3, 4], [0, 1], [6], [5], [3, 7], [7], [8], []]
N = len(adj_list)  # 그래프 정점 수
visited = [False] * N  # 방문되면 True로
s = []

def dfs(v):
    visited[v] = True  # v 방문
    for w in adj_list[v]:  # v에 인접한 각 정점에 대해
        if not visited[w]:
            dfs(w)  # v에 인접한 방문 안 된 정점 순환 호출
    s.append(v)  # i에서 진출하는 간선이 더 이상 없으므로 i를 s에 추가

for i in range(N):
    if not visited[i]:
        dfs(i)

s.reverse()
```

```python

# 결과 출력
print("Topological Sort:")
for v in s:
    print(v, end=' ')
```

```python
import sys

def dijkstra(adj_list, start):
    N = len(adj_list)
    visited = [False] * N
    D = [sys.maxsize] * N
    D[start] = 0
    previous = [None] * N

    for k in range(N):
        m = -1
        min_value = sys.maxsize
        for j in range(N):
            if not visited[j] and D[j] < min_value:
                min_value = D[j]
                m = j
        visited[m] = True

        for w, wt in adj_list[m]:
            if not visited[w]:
                if D[m] + wt < D[w]:
                    D[w] = D[m] + wt
                    previous[w] = m

    return D, previous

```

```python

# 가중치 그래프의 연결 리스트
adj_list = [
    [(1, 2), (2, 1)],
    [(3, 3), (4, 4)],
    [(0, 1), (1, 5)],
    [(6, 6)],
    [(5, 2), (7, 1)],
    [(3, 2), (7, 5)],
    [(7, 3)],
    [(8, 2)],
    [(6, 1)]
]

start = 0

D, previous = dijkstra(adj_list, start)

# 결과 출력
print("Shortest Paths from Vertex", start)
for i in range(len(D)):
    path = []
    current = i
    while current is not None:
        path.append(current)
        current = previous[current]
    path.reverse()
    print("Vertex", i, "- Path:", path, "- Distance:", D[i])
```
