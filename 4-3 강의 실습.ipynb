```python
class BiListNode:
  def __init__(self):
    self.data = None
    self.prev = None
    self.next = None


node0 = BiListNode()
node0.data = 0
head = node0
print(node0.data)

#head(node0)
```

```python
#print Circlar Linked list
def print_Blist(node):
  head = node
  current = node
  while(current.next != None):
    print(current.data,end = " <-> ")
    current = current.next
  print(current.data)
```

```python
dummyHead = BiListNode()
dummyHead.data = "dummy"

head = dummyHead
node0 = BiListNode()
node0.data = 0
node0.prev = dummyHead
dummyHead.next = node0

#head(dummyHead) <-> node0
```

```python
#link node1
node1 = BiListNode()  #init node
node1.data = 10       #set data

node1.prev = node0    #link node node 0 <- node1
node0.next = node1    #link node node 0 -> node1

print(node1.data)
print(node1.prev.data)#node0
```

```python
node2 = BiListNode()
node2.data = 20
node2.prev = node1
node1.next = node2

node3 = BiListNode()
node3.data = 30
node3.prev = node2
node2.next = node3

node4 = BiListNode()
node4.data = 40
node4.prev = node3
node3.next = node4

#head(dummy_head) <-> node0 <-> node1 <-> node2 <-> node3 <-> node4
```

```python
print_Blist(head)
```

```python
#insert 70 between 20 and 30
newItem = 70
prev = node2

newNode = BiListNode()
newNode.data = newItem        #70
newNode.prev = prev           #node2 <- newNode
newNode.next = prev.next      #newNode -> node3
newNode.next.prev = newNode
prev.next = newNode

```

```python
print_Blist(head)
```

```python
#delete 20

current = node2
current.prev.next = current.next      
current.next.prev = current.prev
del(current)
```

```python
print_Blist(head)
```
