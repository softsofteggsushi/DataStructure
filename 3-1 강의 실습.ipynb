3-1 강의 실습

```python
#list insert 50
lst = [0, 10, 20, 30, 40]
print(lst)

print("\nlist inset50\n")
lst.append(50)
print(lst)
```

```python

lst = [0, 10, 20, 30, 40]
print(lst)

print("\nlist inset50 at index 3\n")
lst.insert(3, 50)
print(lst)
```

```python
lst = [0, 10, 20, 30, 40]
print(lst)

print("\nlist delete20 at index 2\n")
del(lst[2])
print(lst)
```

```python
lst = [0, 10, 20, 30, 40]

print (lst[1])            #print index 1
print (lst.index(10))     #print index of data 10
print (len(lst))          #print length list
```
